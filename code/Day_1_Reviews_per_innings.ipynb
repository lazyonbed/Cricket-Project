In a T20 match, there are 2 innings: 1st innings and 2nd innings. What is the share of decisions challenged according to 1st innings and 2nd innings?

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('dark_background') # I love dark mode.
```

```python
# Load the data
drs_data = pd.read_excel("../data/drs_data.xlsx", header=0)
# Print the total number of challenges in the dataset
print("Total number of challenges in the dataset: ", drs_data.shape[0])
```

```python
# Calculate the count of each decision challenged according to which innings it occured in
innings_counts = drs_data['Innings'].value_counts(sort=False)
print("In the 1st innings, umpire decisions were challenged", innings_counts[1], "times, accounting for", round(innings_counts[1]/innings_counts.sum()*100, 2), "% of all challenges.")
print("In the 2nd innings, umpire decisions were challenged", innings_counts[2], "times, accounting for", round(innings_counts[2]/innings_counts.sum()*100, 2), "% of all challenges.")
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
# Create a pie chart
ax.pie(innings_counts, labels=["1st innings", "2nd innings"], autopct='%1.2f%%', colors=["green", "blue"], pctdistance=0.85, startangle=90)
ax.set_title('Review distribution by innings')
# Draw circle
centre_circle = plt.Circle((0,0),0.70,fc='black')
fig.gca().add_artist(centre_circle)
ax.axis("equal")
plt.tight_layout()
plt.savefig("../plots/innings_distribution_reviews.png", bbox_inches='tight', dpi=720)
plt.show()
```
