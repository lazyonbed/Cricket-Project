```python
import pandas as pd # type: ignore
import seaborn as sns # type: ignore
import matplotlib.pyplot as plt # type: ignore

plt.style.use('dark_background') # I love dark mode.
```

```python
# Load the data
drs_data = pd.read_excel("../data/drs_data.xlsx", header=0)
# Print the total number of challenges in the dataset
print("Total number of challenges in the dataset: ", drs_data.shape[0])
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
# Create countplot
ax = sns.countplot(data=drs_data, x="Review By", palette=sns.color_palette("bright"))
ax.set_title("Number of reviews by each team")
ax.set_xlabel("Team")
ax.set_ylabel("Number of reviews")
plt.tight_layout()
plt.savefig("../plots/decision_challenged_teams.png", bbox_inches='tight', dpi=720)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
# Create countplot
ax = sns.countplot(data=drs_data, x="Review By", hue="Result", palette=sns.color_palette("deep"))
ax.set_title("Number of successful/unsuccessful reviews by each team")
ax.set_xlabel("Team")
ax.set_ylabel("Number of reviews")
plt.tight_layout()
plt.savefig("../plots/decision_challenged_teams_result.png", bbox_inches='tight', dpi=720)
plt.show()
```
