```python
import pandas as pd # type: ignore
import seaborn as sns # type: ignore
import matplotlib.pyplot as plt # type: ignore

plt.style.use('dark_background') # I love dark mode.
```

```python
# Load the data
drs_data = pd.read_excel("../data/drs_data.xlsx", header=0)
# Print the total number of challenges in the dataset
print("Total number of challenges in the dataset: ", drs_data.shape[0])
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
# Create countplot
ax = sns.countplot(data=drs_data, x="Umpire", color="green")
ax.set_title("Review distribution by umpire")
# ax.axis("equal")
ax.set_xlabel("Umpire")
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
ax.set_ylabel("Number of umpires decisions challenged")
plt.tight_layout()
plt.savefig("../plots/umpires_distribution_reviews.png", bbox_inches='tight', dpi=720)
plt.show()
```

```python
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(111)
# Create countplot
ax = sns.countplot(data=drs_data, x="Umpire", hue= "Result")
ax.set_title("Review distribution by umpire")
# ax.axis("equal")
ax.set_xlabel("Umpire")
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
ax.set_ylabel("Number of umpires decisions challenged")
plt.tight_layout()
plt.savefig("../plots/umpires_succ_unsucc_distribution_reviews.png", bbox_inches='tight', dpi=720)
plt.show()
```

```python
drs_data["Umpire decision"] = drs_data["Result"].apply(lambda x: "Correct" if x=="Unsuccessful" else "Overturned")
```

```python
drs_data["Umpire decision"].value_counts(sort=False)
```

```python
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(111)
# Create countplot
ax = sns.countplot(data=drs_data, x="Umpire", hue="Umpire decision", palette=["r", "g"])
ax.set_title("IPL 2024 Decision Review System")
# ax.axis("equal")
ax.set_xlabel("Umpire")
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
ax.set_ylabel("Number of decisions challenged")
plt.tight_layout()
plt.savefig("../plots/umpires_correct_overturned_distribution_reviews.png", bbox_inches='tight', dpi=720)
plt.show()
```
