```python
# Import libraries
import pandas as pd # type: ignore
import seaborn as sns # type: ignore
import matplotlib.pyplot as plt # type: ignore

plt.style.use('dark_background')
```

```python
# Load the data
drs_data = pd.read_excel("../data/drs_data.xlsx", header=0)
# Print the total number of challenges in the dataset
print("Total number of challenges in the dataset: ", drs_data.shape[0])
```

```python
# Number of challenges by decision challenged
wide_reviews = drs_data[drs_data["Decision Challenged"]=="Wide"]
wicket_reviews = drs_data[drs_data["Decision Challenged"]=="Wicket"]
noball_reviews = drs_data[drs_data["Decision Challenged"]=="NoBall"]
```

```python
print("Wides have been reviewed", wide_reviews.__len__(), "times.")
print("Wickets have been reviewed", wicket_reviews.__len__(), "times.")
print("No Balls have been reviewed", noball_reviews.__len__(), "times.")
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
# Create a pie chart
ax.pie(drs_data["Decision Challenged"].value_counts(sort=True),
       labels=drs_data["Decision Challenged"].value_counts(sort=True).index.to_list(), # wanted to automatically get the labels
       autopct='%1.2f%%',
       colors=["green", "blue", "red"],
       pctdistance=0.85,
       startangle=45)
ax.set_title('Types of reviews')
# Draw circle
centre_circle = plt.Circle((0,0),0.70,fc='black')
fig.gca().add_artist(centre_circle)
ax.axis("equal")
plt.tight_layout()
plt.savefig("../plots/piechart_distribution_reviews.png", bbox_inches='tight', dpi=720)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
# Create countplot
ax = sns.countplot(data=drs_data, x="Decision Challenged", palette=sns.color_palette("Set2"))
ax.set_title("Types of reviews")
ax.set_xlabel("Review type")
ax.set_ylabel("Number of reviews")
plt.tight_layout()
plt.savefig("../plots/decision_challenged_distribution.png", bbox_inches='tight', dpi=720)
plt.show()
```
